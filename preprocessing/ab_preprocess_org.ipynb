```python
import dask.dataframe as pd
import numpy as np
from decimal import Decimal
```

```python
df = pd.read_csv("complex_org_m5.csv").compute()
```

```python
df
```

```python
anticlock_orientation = 'ABCDABCD'*2
clock_orientation = "".join(reversed(clock_orientation))

def find_orientation(sequence):
    if sequence in clock_orientation:
        return 'clockwise'
    elif sequence in anticlock_orientation:
        return 'anticlockwise'
    else:
        return 'random'
```

```python
df['orientation'] = df['sequence'].apply(find_orientation)
# df[(df['orientation']=='clockwise')]
```
